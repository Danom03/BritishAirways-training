# Web Scraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
def extract_data():
    head = []
    user_status = []
    content = []
    for i in range(1,367):
        # Extracts all reviews page hyperlinks
        url = "https://www.airlinequality.com/airline-reviews/british-airways/page/{page}/".format(page=i)
        r = requests.get(url)       # Requesting access to the URL
        html_doc = r.text           # Converting html to text
        soup = BeautifulSoup(html_doc,'html.parser')  #Scraper
        
        # Main Heading of the Review
        for i in soup.find_all(class_ = 'text_header'):
            head.append(i.text)
        
        # User information    
        for i in soup.find_all(class_ = 'userStatusWrapper'):
            user_status.append(i.text)
        
        # Content of the Review
        for i in soup.find_all(class_ = 'text_content'):
            content.append(i.text)
    return head, user_status, content
```

```python
head, user_status, content = extract_data()
```

# Saving scraped data to .csv File

```python
import pandas as pd
```

###### Creating pandas DataFrame from data

```python
df = pd.DataFrame(data = [user_status, head, content])
```

```python
# Transposing to swap rows and columns
df_new = df.T
```

```python
# Naming columns
df_new.columns = ['user_status', 'heading', 'reviews']
```
